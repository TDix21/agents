`ollama ls` lists all the models you've downloaded  
`ollama rm <model_name>` deletes the specified model from your downloads

```python
ollama = OpenAI(base_url='http://localhost:11434/v1', api_key='ollama')
model_name = "llama3.2"

response = ollama.chat.completions.create(model=model_name, messages=messages)
answer = response.choices[0].message.content

display(Markdown(answer))
competitors.append(model_name)
answers.append(answer)
```

```python
# So where are we?

print(competitors)
print(answers)
```

```python
# It's nice to know how to use "zip"
for competitor, answer in zip(competitors, answers):
    print(f"Competitor: {competitor}\n\n{answer}")
```

```python
# Let's bring this together - note the use of "enumerate"

together = ""
for index, answer in enumerate(answers):
    together += f"# Response from competitor {index+1}\n\n"
    together += answer + "\n\n"
```

```python
print(together)
```

```python
judge = f"""You are judging a competition between {len(competitors)} competitors.
Each model has been given this question:

{question}

Your job is to evaluate each response for clarity and strength of argument, and rank them in order of best to worst.
Respond with JSON, and only JSON, with the following format:
{{"results": ["best competitor number", "second best competitor number", "third best competitor number", ...]}}

Here are the responses from each competitor:

{together}

Now respond with the JSON with the ranked order of the competitors, nothing else. Do not include markdown formatting or code blocks."""
```

```python
print(judge)
```

```python
judge_messages = [{"role": "user", "content": judge}]
```

```python
# Judgement time!
response = openrouter.chat.completions.create(
    model="gpt-5-mini",
    messages=judge_messages,
)
results = response.choices[0].message.content
print(results)
```

```python
# OK let's turn this into results!

results_dict = json.loads(results)
ranks = results_dict["results"]
for index, result in enumerate(ranks):
    competitor = competitors[int(result)-1]
    print(f"Rank {index+1}: {competitor}")
```

```python
models = ["anthropic/claude-opus-4-6", "anthropic/claude-opus-4-6", "google/gemini-3-pro-preview", "anthropic/claude-opus-4-5"]
llm_responses =[]

ai_ml = OpenAI(
    base_url="https://api.aimlapi.com/v1",
    api_key=ai_ml_api_key
)



messages = [{'role': 'user', 'content': 'Please generate a complex question that would take many LLMs to answer.'}]

response = ai_ml.chat.completions.create(
    model=models[0],
    messages=messages
)

hard_question = response.choices[0].message.content

print(f"THIS IS THE QUESTION: {hard_question}")

messages = [{'role': 'user', 'content': f'You\'ve been given an incredibly hard question, and have 2 other LLMS at your disposal as helpers: {models[2]} and {models[3]}. Along with these 2 models, please divide this question into smaller tasks that you and the models can address. Highlight the section that you want address. Answer that portion and then have the others models address their portion. Here is the question: {hard_question}'}]

response = ai_ml.chat.completions.create(
    model=models[1],
    messages=messages
)

opus_question_split_into_tasks = response.choices[0].message.content

print(f"This is Opus 4.6's response: {opus_question_split_into_tasks}")

llm_responses.append(opus_question_split_into_tasks)

messages = [{'role': 'user', 'content': f'You are working with 2 other models: {models[1]} and {models[3]} to answer this question: {hard_question}. {models[1]} has split the question into separate tasks and it has highlighted a portion of the problem it wants you to deal with. Here it is: {opus_question_split_into_tasks}. PLEASE DO NOT ANSWER ANY OTHER PORTION OUTSIDE OF THE ONE THAT {models[1]} ASSIGNED YOU TO ANSWER!'}]

response = ai_ml.chat.completions.create(
    model=models[2],
    messages=messages
)

gemini_response = response.choices[0].message.content

print(f"This is Gemini 3 Pro Preview's response: {gemini_response}")

llm_responses.append(gemini_response)

messages = [{'role': 'user', 'content': f'You are working with 2 other models: {models[1]} and {models[2]} to answer this question: {hard_question}. {models[1]} and {models[2]} have answered their portion and {models[1]} has split the question into separate tasks and it has highlighted a portion of the problem it wants you to deal with. Here it is: {opus_question_split_into_tasks}. PLEASE DO NOT ANSWER ANY OTHER PORTION OUTSIDE OF THE ONE THAT {models[1]} ASSIGNED YOU TO ANSWER!'}]

response = ai_ml.chat.completions.create(
    model=models[3],
    messages=messages
)

opus_response = response.choices[0].message.content

print(f"This is Opus 4.5's response: {opus_response}")

llm_responses.append(opus_response)

messages=[{'role': 'user', 'content': f'I asked a LLM to create such a hard question that multiple LLMS have to come together to answer a question. {models[0]} created this question: {hard_question}. I then asked 3 LLMS: {models[1:3]} to help answer. {models[1]} answered their portion of the question and passed it to {models[2]} and {models[3]}. {models[1]} answered with {opus_question_split_into_tasks}. {models[2]} responded with {gemini_response} and {models[3]} responded with {opus_response}. Please aggregate their responses and highlight which model answered which portion. Thank you.'}]

response = ai_ml.chat.completions.create(
    model=models[1],
    messages=messages
)

final_aggregation = response.choices[0].message.content

print(final_aggregation)
```
