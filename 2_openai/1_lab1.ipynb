```python
# The imports

from dotenv import load_dotenv
from agents import Agent, Runner, trace
from openai import OpenAI
import os

os.environ["OPENAI_API_KEY"] = os.getenv("PUSHOVER_USER")
```

```python
# The usual starting point

load_dotenv(override=True)

import asyncio
from openai import AsyncOpenAI
from agents import Agent, Runner, function_tool, set_default_openai_client
from agents.models.openai_chatcompletions import OpenAIChatCompletionsModel

# 1. Configure OpenRouter client globally
client = AsyncOpenAI(base_url="https://openrouter.ai/api/v1", api_key="OPENROUTER_API_KEY")
set_default_openai_client(client)

# 2. Build a simple agent
@function_tool
def get_time() -> str:
    from datetime import datetime
    return f"It's {datetime.utcnow().isoformat()} UTC"

agent = Agent(
    name="TimeAgent",
    instructions="Answer only with the current UTC time when asked.",
    model=OpenAIChatCompletionsModel(model="gpt-4", openai_client=client),  # or any model
    tools=[get_time]
)

# 3. Run the agent
result = await Runner.run(agent, "What time is it?")
print("Agent says:", result.final_output)
```

```python

# Make an agent with name, instructions, model
agent = Agent(name="Jokester", instructions="You are a joke teller", model="gpt-4o-mini")
```

```python
# Run the joke with Runner.run(agent, prompt) then print final_output

with trace("Telling a joke"):
    result = await Runner.run(agent, "Tell a joke about Autonomous AI Agents")
    print(result.final_output)
```
